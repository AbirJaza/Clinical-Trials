```python
import csv
import pandas as pd
```

```python
'''
1) inject the secondary_id column into the base tables that do not contain it
   There are 4 tables : China, brazil, iran and peru 
2) take only the two colums of each table (trial_id and sec_id) 
3) merge sec_ids columns of the same table (to have only 2 columns trial_id and sec_id)
4) remove duplicates if they exist

'''
```

```python
### USA

df_ct_main = pd.read_csv('../Data/USA/ct_main.csv', sep='|')
data_USA = df_ct_main[['NCT_ID','SECONDARY_ID']]  # len(data_USA) = 268859
# Determining duplicates
mask = data_USA.duplicated(keep=False)
double = data_USA[mask]   #len(double) = 0 ==> no duplicates in data USA
data_USA = data_USA.rename(columns={ data_USA.columns[0]: "id" , data_USA.columns[1]: "sec_id"})
data_USA[0:4]
```

```python
### Thailand

df_tctr_main = pd.ExcelFile('../Data/Thailand/tctr_main.xlsx').parse('tctr_main')
data_Thailand = df_tctr_main.reindex(columns=['TRIAL_ID','SECONDARY_ID']) # len(data_Thailand) = 1838
# Determining duplicates
mask = data_Thailand.duplicated(keep=False)
double = data_Thailand[mask]   #len(double) = 0 ==> no duplicates in data Thailand
data_Thailand = data_Thailand.rename(columns={ data_Thailand.columns[0]: "id" , data_Thailand.columns[1]: "sec_id"})
data_Thailand[0:4]
```

```python
### PANAFRICA

df_pactr_trials_clean = pd.ExcelFile('../Data/PANAFRICA/pactr_trials_clean.xlsx').parse('Sheet1')
data_PANAFRICA = df_pactr_trials_clean[['id','sec_ids']] # len(data_PANAFRICA) = 2260
# Determining duplicates
mask = data_PANAFRICA.duplicated(keep=False)
double = data_PANAFRICA[mask]   #len(double) = 0 ==> no duplicates in data PANAFRICA
data_PANAFRICA = data_PANAFRICA.rename(columns={ data_PANAFRICA.columns[0]: "id" , data_PANAFRICA.columns[1]: "sec_id"})
data_PANAFRICA[0:4]
```

```python
### Peru
df_secondary_id = pd.read_csv('../Data/Peru/secondary_id.csv', sep='|')
data_Peru = df_secondary_id[['TRIAL_ID','SECONDARY_ID']] # len(data_Peru) = 7028
# Determining duplicates
mask = data_Peru.duplicated(keep=False)
double = data_Peru[mask]   #len(double) = 5062 ==> len(data Peru) = 1966
data_Peru = data_Peru.drop_duplicates(keep=False)
data_Peru = data_Peru.rename(columns={ data_Peru.columns[0]: "id" , data_Peru.columns[1]: "sec_id"})
data_Peru[0:4]
```

```python
### Iran

df_sec_ids = pd.read_csv('../Data/Iranian RCT/sec_ids.csv', sep='|')
data_Iran = df_sec_ids[['TRIAL_ID','SECONDARY_ID']] # len(data_Iran) = 7028
# Determining duplicates
mask = data_Iran.duplicated(keep=False)
double = data_Iran[mask]   #len(double) = 52 ==> len(data Iran) = 17018
data_Iran = data_Iran.drop_duplicates(keep=False)
data_Iran = data_Iran.rename(columns={ data_Iran.columns[0]: "id" , data_Iran.columns[1]: "sec_id"})
data_Iran[0:4] 
```

```python
# China

df_ChiCTR_sec_id = pd.read_csv('../Data/China/ChiCTR_sec_id.csv', sep='|')
data_China = df_ChiCTR_sec_id[['TRIAL_ID','SECONDARY_ID']]  # len(data_China) = 15651
# Determining duplicates
mask = data_China.duplicated(keep=False)
double = data_China[mask]   #len(double) = 3 ==> len(data China) = 15648
data_China = data_China.drop_duplicates(keep=False)
data_China = data_China.rename(columns={ data_China.columns[0]: "id" , data_China.columns[1]: "sec_id"})
data_China[0:4] 
```

```python
### Brazil

df_brazil_secondaryid = pd.read_csv('../Data/Brazil/brazil_secondaryid.csv', sep='|')
data_Brazil = df_brazil_secondaryid.reindex(columns=['TRIAL_ID','SEC_ID'])  # len(data_Brazil) = 3807
# Determining duplicates
mask = data_Brazil.duplicated(keep=False)
double = data_Brazil[mask]   #len(double) = 6 ==> len(data China) = 3801
data_Brazil = data_Brazil.drop_duplicates(keep=False)
data_Brazil = data_Brazil.rename(columns={ data_Brazil.columns[0]: "id" , data_Brazil.columns[1]: "sec_id"})
data_Brazil[0:4] 
```

```python
### Srilanka Slctr

df_Slctr = pd.ExcelFile('../Data/Slctr/Slctr_trials.xlsx').parse('Sheet1')
all_data_Srilanka = df_Slctr[['trial_id','uni_trial_no','trial_no_others']]  
df1 = all_data_Srilanka[['trial_id','uni_trial_no']]
df1 = df1.rename(columns={ df1.columns[0]: "id" , df1.columns[1]: "sec_id"})
df2 = all_data_Srilanka[['trial_id','trial_no_others']]
df2 = df2.rename(columns={ df2.columns[0]: "id" , df2.columns[1]: "sec_id"})
data_Srilanka = df1.append(df2)   # len(data_Srilanka) = 536
# Determining duplicates
mask = data_Srilanka.duplicated(keep=False)
double = data_Srilanka[mask]   #len(double) = 60 ==> len(data Srilanka) = 476
data_Srilanka = data_Srilanka.drop_duplicates(keep=False)
data_Srilanka[0:4] 
```

```python
# korea CRIS
df_cris_trials = pd.ExcelFile('../Data/CRIS/cris_trials.xlsx').parse('Sheet1')
all_data_korea = df_cris_trials[['cris_id','registry_name']]

s = list(all_data_korea['registry_name'])
new_s = []
for i in s:
    l = str(i).split('-')
    if len(l)!=1:
        new_s.append(l[1])
    else:
        new_s.append(i)
        
se = pd.Series(new_s)
all_data_korea['sec_id'] = se.values  
data_korea = all_data_korea.rename(columns={ all_data_korea.columns[0]: "id" })
data_korea = data_korea[['id','sec_id']]  #len(data_korea) = 2811
# Determining duplicates
mask = data_korea.duplicated(keep=False)
double = data_korea[mask]   #len(double) = 0 ==> len(data Srilanka) = 2811
data_korea = data_korea.drop_duplicates(keep=False)
data_korea[0:4] 
```

```python
# Japan 
df_ctr_data_e = pd.read_csv('../Data/Japan/ctr_data_e.csv', sep=',')
data_japan = df_ctr_data_e.reindex(columns=['Unique ID issued by UMIN','Study ID_1','Study ID_2'])
df1 = data_japan[['Unique ID issued by UMIN','Study ID_1']]
df1 = df1.rename(columns={ df1.columns[0]: "id" , df1.columns[1]: "sec_id"})
df2 = data_japan[['Unique ID issued by UMIN','Study ID_2']]
df2 = df2.rename(columns={ df2.columns[0]: "id" , df2.columns[1]: "sec_id"})
data_japan = df1.append(df2)   # len(data_japan) = 63468
# Determining duplicates
mask = data_japan.duplicated(keep=False)
double = data_japan[mask]   #len(double) = 62098 ==> len(data japan) = 1370
data_japan = data_japan.drop_duplicates(keep=False)
data_japan[0:4] 
```

```python
# ICTRP
df_ictrp_data_2018 = pd.read_csv('../Data/ICTRP/ictrp_data_2018.csv', sep='|')
all_data_ICTRP = df_ictrp_data_2018.reindex(columns=['TrialID','SecondaryID1','SecondaryID2','SecondaryID3','SecondaryID4','SecondaryID5'])
df1 = all_data_ICTRP[['TrialID','SecondaryID1']].rename(columns={ all_data_ICTRP.columns[0]: "id" , all_data_ICTRP.columns[1]: "sec_id"})
df2 = all_data_ICTRP[['TrialID','SecondaryID2']].rename(columns={ all_data_ICTRP.columns[0]: "id" , all_data_ICTRP.columns[2]: "sec_id"})
df3 = all_data_ICTRP[['TrialID','SecondaryID3']].rename(columns={ all_data_ICTRP.columns[0]: "id" , all_data_ICTRP.columns[3]: "sec_id"})
df4 = all_data_ICTRP[['TrialID','SecondaryID4']].rename(columns={ all_data_ICTRP.columns[0]: "id" , all_data_ICTRP.columns[4]: "sec_id"})
df5 = all_data_ICTRP[['TrialID','SecondaryID5']].rename(columns={ all_data_ICTRP.columns[0]: "id" , all_data_ICTRP.columns[5]: "sec_id"})
data_ICTRP = df1.append([df2,df3,df4,df5])  # len(data_ICTRP) = 570730
# Determining duplicates
mask = data_ICTRP.duplicated(keep=False)
double = data_ICTRP[mask]   #len(double) = 513695 ==> len(data japan) = 57035
data_ICTRP = data_ICTRP.drop_duplicates(keep=False)
data_ICTRP[0:4] 
```

```python
# EU
df_eucrt_data = pd.read_csv('../Data/EU/eucrt_data.csv', sep='|')
all_data_EU = df_eucrt_data.reindex(columns=['EudraCT Number:','A.5.1 ISRCTN(International Standard Randomised Control Trial) number:','A.5.2 US NCT (ClinicalTrials.gov registry) number:'])
df1 = all_data_EU[['EudraCT Number:','A.5.1 ISRCTN(International Standard Randomised Control Trial) number:']].rename(columns={ all_data_EU.columns[0]: "id" , all_data_EU.columns[1]: "sec_id"})
df2 = all_data_EU[['EudraCT Number:','A.5.2 US NCT (ClinicalTrials.gov registry) number:']].rename(columns={ all_data_EU.columns[0]: "id" , all_data_EU.columns[2]: "sec_id"})
data_EU = df1.append([df2])  # len(data_IEU) = 64474
# Determining duplicates
mask = data_EU.duplicated(keep=False)
double = data_EU[mask]   #len(double) = 59142 ==> len(data EU) = 5332
data_EU = data_EU.drop_duplicates(keep=False)
data_EU[0:4] 
```

```python
# ANZCTR
df_anzctr_main = pd.read_csv('../Data/ANZCTR/anzctr_main.csv', sep='|')
all_data_anzctr = df_anzctr_main.reindex(columns=['TRIAL_ID','SECONDARY_ID1','SECONDARY_ID2','SECONDARY_ID3'])
df1 = all_data_anzctr.reindex(columns=['TRIAL_ID','SECONDARY_ID1']).rename(columns={ all_data_anzctr.columns[0]: "id" , all_data_anzctr.columns[1]: "sec_id"})
df2 = all_data_anzctr.reindex(columns=['TRIAL_ID','SECONDARY_ID2']).rename(columns={ all_data_anzctr.columns[0]: "id" , all_data_anzctr.columns[2]: "sec_id"})
df3 = all_data_anzctr.reindex(columns=['TRIAL_ID','SECONDARY_ID3']).rename(columns={ all_data_anzctr.columns[0]: "id" , all_data_anzctr.columns[3]: "sec_id"})
data_anzctr = df1.append([df2,df3])  # len(data_ICTRP) = 64263
# Determining duplicates
mask = data_anzctr.duplicated(keep=False)
double = data_anzctr[mask]   #len(double) = 39439 ==> len(data ANZCTR) = 24824
data_anzctr = data_anzctr.drop_duplicates(keep=False)

def firstElement(x):
    if type(x)!=float:
        x = x[2:-2]
    return(x)
        
new_sec_id = data_anzctr['sec_id'].apply(firstElement)
data_anzctr.insert(loc=0, column='sec_id1', value=new_sec_id)
data_anzctr = data_anzctr[['id','sec_id1']].rename(columns={ 'sec_id1': "sec_id"})
data_anzctr[0:4] 
```
