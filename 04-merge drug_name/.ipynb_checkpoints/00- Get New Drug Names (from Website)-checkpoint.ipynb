```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from lxml import html
import time 
import pickle
import requests
import re
```

```python
def getPage(link): 
    response = requests.get(link)
    content_source = response.text.encode("utf-8")
    xml_doc = html.document_fromstring(content_source)
    print(content_source)
    return(xml_doc)
```

### 01- GET DRUG NAMES AND SYNONYMS

```python
driver = webdriver.Chrome()
Project_URL = "https://druginfo.nlm.nih.gov"
```

```python
drugs_dict = dict()

for page_num in list(map(chr, range(97, 123))):     #Parcourir toutes les lettres alphabets 
    
    URL = Project_URL + '/drugportal/drug/names/' + page_num
    xml_doc = getPage(URL)
    drug_links = xml_doc.xpath('//tr/td/a/@href')
    
    for drug_link in drug_links[1:len(drug_links)]:    
        
        drug_name = drug_link.replace('/drugportal/name/','').replace('%20', ' ')
        print(drug_name)
        
        URL_drug = Project_URL + drug_link
        driver.get(URL_drug)
        time.sleep(3)
        
        try:
            
            check = driver.find_element_by_xpath("//span[contains(text(),'1 result for Name/Synonym equals')]")
            URL_drug = Project_URL + drug_link
            driver.get(URL_drug)
            time.sleep(3)
            
            try:
                drug_synonyms = driver.find_element_by_xpath("//tr/td/button")
                drug_synonyms = drug_synonyms.get_attribute("onclick")
                elements = re.findall('<li>(.*)<br>', drug_synonyms)
                drugs_dict[drug_name] = elements[0].split('<br><li>')
                
            except:
                print("NO SYNONYMS FOR THIS DRUG:", drug_name)
            
        except:
            
            #MORE THAN ONE RESULT 
            list_drugs_links = []
            list_global_drug_names = []
            sub_drug_links = driver.find_elements_by_xpath("//a[@class='link']")

            for sub_drug_link in sub_drug_links:
                sub_drug_link_href = sub_drug_link.get_attribute("href")
                list_drugs_links.append(sub_drug_link_href)
                list_global_drug_names.append(sub_drug_link.text)   
            print(list_drugs_links)                  

            for link,name in zip(list_drugs_links,list_global_drug_names):
                driver.get(link)
                time.sleep(3)
                try:
                    drug_synonyms = driver.find_element_by_xpath("//tr/td/button")
                    drug_synonyms = drug_synonyms.get_attribute("onclick")
                    elements = re.findall('<li>(.*)<br>', drug_synonyms)
                    drugs_dict[name] = elements[0].split('<br><li>')
                    
                except:
                    print("NO DRUG SYNONYMS FOR THIS DRUG:",link)
```

```python
#GET DRUG NAMES FROM DICT
all_drugs = []
rec_num = 0
for key, values in drugs_dict.items():
    rec_num += 1
    all_drugs.append(tuple((rec_num,key.replace('/drugportal/name/','').replace('%20', ' '))))
    for value in values:
        all_drugs.append(tuple((rec_num,key,value)))
```

### 02- GET Term Drug w/ Synonyms in "See search details"

```python
driver = webdriver.Chrome()
Project_URL = "https://druginfo.nlm.nih.gov"

new_drugs_dict = []
#count = 0
#for page_num in list(map(chr, range(97, 123))):     #Parcourir toutes les lettres alphabets 
#for page_num in list(map(chr, range(98, 123))):  
for page_num in list('a'):
    URL = Project_URL + '/drugportal/drug/names/' + page_num
    xml_doc = getPage(URL)
    drug_links = xml_doc.xpath('//tr/td/a/@href')    #Parcourir la liste des Drug Names
    
    for drug_link in drug_links[1:len(drug_links)]:    
        global_drug_name = drug_link.replace('/drugportal/name/','').replace('%20', ' ')
        URL_drug = Project_URL + drug_link
        driver.get(URL_drug)
        time.sleep(3)
        
        try:
            #count +=1
            check = driver.find_element_by_xpath("//span[contains(text(),'1 result for Name/Synonym equals')]")
            print(global_drug_name)
            
            try:
                #For Clinical trials (ClinicalTrials.gov)
                trial_link = driver.find_elements_by_xpath("//a[contains(@onmouseover, '\<b\>ClinicalTrials.gov\</b\>')]")
                trial_link = trial_link[0].get_attribute("href")
                driver.get(trial_link)
                time.sleep(3)    
               
                #FOR SEARCH DETAILS
                search_details = driver.find_element_by_xpath("//a[@id='search-result-detail-tab']")
                search_details_link = search_details.get_attribute("href")
                driver.get(search_details_link)
                try:
                    search_terms = driver.find_elements_by_xpath("//table[@class='w3-table']/tbody/tr/td[1]")
                    total_studies_in_search_res = driver.find_elements_by_xpath("//table[@class='w3-table']/tbody/tr/td[2]")
                    #print(total_studies_in_search_res)
                    total_studies_in_entire_db = driver.find_elements_by_xpath("//table/tbody/tr/td[3]")
                    #print(total_studies_in_entire_db)

                    #SAVE RESULTS IN A LIST
                    for search_term, studies_in_search_res, studies_in_entire_db in zip(search_terms,total_studies_in_search_res,total_studies_in_entire_db):
                        try:
                            'vertical-align' in search_term.get_property("style")[2]
                        except :
                            new_drugs_dict.append(tuple((global_drug_name,
                                                         global_drug_name,
                                                         search_term.text,
                                                         "0", 
                                                         studies_in_search_res.text, 
                                                         studies_in_entire_db.text)))
                        else:
                            new_drugs_dict.append(tuple((global_drug_name,
                                                         global_drug_name,
                                                         search_term.text,
                                                         "1", 
                                                         studies_in_search_res.text, 
                                                         studies_in_entire_db.text)))
                            
                except:
                    print("NO SEARCH DETAILS FOR",global_drug_name,", LINK:",search_details_link.text)
                    
                    
            except:
                print("haha NO TRIAL LINK FOR",global_drug_name,", LINK:",URL_drug)
                
        except:
            
            #MORE THAN ONE RESULT  
            list_drugs_links = []
            list_global_drug_names = []
            sub_drug_links = driver.find_elements_by_xpath("//a[@class='link']")
            
            for sub_drug_link in sub_drug_links:
                sub_drug_link_href = sub_drug_link.get_attribute("href")
                list_drugs_links.append(sub_drug_link_href)
                list_global_drug_names.append(sub_drug_link.text)   

            print(list_drugs_links)                  

            for link,name in zip(list_drugs_links,list_global_drug_names):

                driver.get(link)
                time.sleep(3)

                try:
                    #FOR CLINICAL TRIALS (ClinicalTrials.gov)
                    trial_link = driver.find_element_by_xpath("//a[contains(@onmouseover,'\<b\>ClinicalTrials.gov\</b\>')]")
                    trial_link = trial_link.get_attribute("href")
                    print(trial_link)
                    driver.get(trial_link)
                    time.sleep(3)

                    #FOR SEARCH DETAILS
                    search_details = driver.find_element_by_xpath("//a[@id='search-result-detail-tab']")
                    search_details_link = search_details.get_attribute("href")
                    driver.get(search_details_link)
                    time.sleep(3)

                    search_terms = driver.find_elements_by_xpath("//table[@class='w3-table']/tbody/tr/td[1]")
                    total_studies_in_search_res = driver.find_elements_by_xpath("//table[@class='w3-table']/tbody/tr/td[2]")
                    total_studies_in_entire_db = driver.find_elements_by_xpath("//table/tbody/tr/td[3]")



                    #SAVE RESULTS IN A LIST
                    for search_term, studies_in_search_res, studies_in_entire_db in zip(search_terms,total_studies_in_search_res,total_studies_in_entire_db):
                        try:
                            'vertical-align' in search_term.get_property("style")[2]
                        except :
                            new_drugs_dict.append(tuple((global_drug_name,
                                                         name,
                                                         search_term.text,
                                                         "0", 
                                                         studies_in_search_res.text, 
                                                         studies_in_entire_db.text)))
                        else:
                            new_drugs_dict.append(tuple((global_drug_name,
                                                         name,
                                                         search_term.text,
                                                         "1", 
                                                         studies_in_search_res.text, 
                                                         studies_in_entire_db.text)))
                            
                except:
                    print("NO TRIAL LINK FOR",name,", LINK:",link)
        
print(len(new_drugs_dict))     
```

### 03- GET all Drugs info in trials

```python
driver = webdriver.Chrome()
Project_URL = "https://druginfo.nlm.nih.gov"

new_drugs_dict = []
drug_inf = []

#for page_num in list(map(chr, range(97, 123))):     #Parcourir toutes les lettres alphabets 
for page_num in list('a'):
    URL = Project_URL + '/drugportal/drug/names/' + page_num
    xml_doc = getPage(URL)
    drug_links = xml_doc.xpath('//tr/td/a/@href')    #Parcourir la liste des Drug Names
    
    for drug_link in drug_links[1:len(drug_links)]:    
        global_drug_name = drug_link.replace('/drugportal/name/','').replace('%20', ' ')
        URL_drug = Project_URL + drug_link
        driver.get(URL_drug)
        time.sleep(3)
        
        try:
            #count +=1
            check = driver.find_element_by_xpath("//span[contains(text(),'1 result for Name/Synonym equals')]")
            print(global_drug_name)
            
            try:
                #For Clinical trials (ClinicalTrials.gov)
                trial_link = driver.find_elements_by_xpath("//a[contains(@onmouseover, '\<b\>ClinicalTrials.gov\</b\>')]")
                trial_link = trial_link[0].get_attribute("href")
                driver.get(trial_link)
                time.sleep(3)    
                
                details = driver.find_element_by_class_name("ct-inner_content_wrapper")
                details = details.text.replace('\n',' ; ').replace(' See Search Details','')
                drug_inf.append(tuple((global_drug_name,global_drug_name,details)))               

                while True:
                    
                    time.sleep(3)
                    trial_ids = driver.find_elements_by_xpath("//tbody/tr/td[7]")
                    interventions = driver.find_elements_by_xpath("//tbody/tr/td[6]")

                    for trial_id,intervention in zip(trial_ids,interventions):
                        new_drugs_dict.append(tuple((global_drug_name,global_drug_name,trial_id.text,intervention.text.replace('\n',' ; '))))
                    #print(len(drugs_dict)) 
                        
                    try:
                        elm = driver.find_element_by_link_text(">")
                    except:
                        print("NO PAGINATION IN THIS PAGE, LINK:",trial_link)
                        break;
                    else:
                        if 'paginate_button next disabled' in elm.get_attribute('class'):
                            break;
                        elm.click()
                        time.sleep(3)            

                    
                    
            except:
                print("hana NO TRIAL LINK FOR",global_drug_name,", LINK:",URL_drug)
                
        except:
            
            #MORE THAN ONE RESULT  
            list_drugs_links = []
            list_global_drug_names = []
            sub_drug_links = driver.find_elements_by_xpath("//a[@class='link']")
            
            for sub_drug_link in sub_drug_links:
                sub_drug_link_href = sub_drug_link.get_attribute("href")
                list_drugs_links.append(sub_drug_link_href)
                list_global_drug_names.append(sub_drug_link.text)   

            print(list_drugs_links)                  

            for link,name in zip(list_drugs_links,list_global_drug_names):

                driver.get(link)
                time.sleep(3)

                try:
                    #FOR CLINICAL TRIALS (ClinicalTrials.gov)
                    trial_link = driver.find_element_by_xpath("//a[contains(@onmouseover,'\<b\>ClinicalTrials.gov\</b\>')]")
                    trial_link = trial_link.get_attribute("href")
                    print(trial_link)
                    driver.get(trial_link)
                    time.sleep(3)
                                                
                    details = driver.find_element_by_class_name("ct-inner_content_wrapper")
                    details = details.text.replace('\n',' ; ').replace(' See Search Details','')
                    drug_inf.append(tuple((global_drug_name,name,details)))

                    while True:
                        
                        trial_ids = driver.find_elements_by_xpath("//tbody/tr/td[7]")
                        interventions = driver.find_elements_by_xpath("//tbody/tr/td[6]")
                        
                        for trial_id,intervention in zip(trial_ids,interventions):
                            new_drugs_dict.append(tuple((global_drug_name,name,trial_id.text,intervention.text.replace('\n',' ; '))))
                        #print(len(drugs_dict)) 
                        
                        try:
                            elm = driver.find_element_by_link_text(">")
                        except:
                            print("NO PAGINATION IN THIS PAGE, LINK:",trial_link)
                            break;
                        else:
                            if 'paginate_button next disabled' in elm.get_attribute('class'):
                                break;
                            elm.click()
                            time.sleep(3)                     
                    
                       
                        
                except:
                    print("NO TRIAL LINK FOR",name,", LINK:",link)
        
print('len(new_drugs_dict)=',len(new_drugs_dict))    
print('len(drug_inf)=',len(drug_inf))    
```
