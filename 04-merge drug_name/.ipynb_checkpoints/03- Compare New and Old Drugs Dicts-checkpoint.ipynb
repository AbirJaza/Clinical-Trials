```python
import pandas as pd
import csv
```

### OLD DRUGS DICT (from website)

```python
df_old_drugs = pd.read_excel("../../Data/drugs/old_drug_names_final.xlsx")
df_old_drugs['record_no'] = 'od' + df_old_drugs['record_no'].astype(str)
df_old_drugs = df_old_drugs.drop(columns=['Unnamed: 0'])
df_old_drugs.head()
```

```python
len(df_old_drugs)
```

### NEW DRUGS DICT

```python
df_new_drugs = pd.read_excel('../../Data/drugs/new_drug_names_final.xlsx')
df_new_drugs = df_new_drugs.drop(columns=['Unnamed: 0'])
df_new_drugs.head()
```

```python
len(df_new_drugs)
```

```python
df_old_drugs[df_old_drugs.drug_name == "Condition"]
```

# TEST 

```python
df_new_drugs = df_new_drugs.assign(new_in_old_drugs=df_new_drugs.search_string.isin(df_old_drugs.drug_name).astype(int))
df_new_drugs.head()
```

```python
len(df_new_drugs[df_new_drugs.new_in_old_drugs==1])
```

```python
# Duplicates in old drugs dict
df_dup_old_drugs = df_old_drugs[df_old_drugs.duplicated(['drug_name_lower'])]
len(df_dup_old_drugs)
```

```python
df_dup_old_drugs.head()
```

```python
df_no_dup_old_drugs = df_old_drugs.drop_duplicates(subset='drug_name_lower') 
len(df_no_dup_old_drugs)
```

```python
df_no_dup_old_drugs.to_excel("../../Data/drugs/old_drug_names_final_no_duplicates.xlsx")
```

```python
df_dup_old_drugs.groupby(['record_no']).count()
```
