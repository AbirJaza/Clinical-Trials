## RUN ON ALL DATA

```python
drugs_dict = dict()
trial_count = 0
l_final = []
for filename in Path('../AllPublicXML/').glob('**/*.xml'):
    l_drugs = []
    trial_count += 1
    #print(filename) 
    tree = ET.parse(filename)
    root = tree.getroot()    
    
    #GET TRAIL INFOS
    for trail in root.findall('id_info'):
        trail_id = trail.find('nct_id').text
        #print(trail_id)
        
    #GET INTERVENTION INFOS
    for intervention in root.findall('intervention'):
        intervention_type = intervention.find('intervention_type').text
        intervention_name = intervention.find('intervention_name').text
        l_drugs.append(intervention_name+' ('+intervention_type+')')
        #print(intervention_name,',',intervention_type)
    
    drugs_dict[trail_id] = l_drugs
    #print('###################')
```

```python
#Total of Trials
print(trial_count)
```
